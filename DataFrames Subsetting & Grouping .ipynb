```python
# Add century to the year and convert to datetime object. 
def make_date(date_str):
    # Remove the slash
    m, d, y = date_str.split('/')
    m = int(m)
    d = int(d)
    y = int(y)
    if y > 13:
        y += 1900
    else:
        y += 2000
    return dt.datetime(y, m, d)
    
```

# Grouping

```python
# Split, Apply and Combine
# Split the data into groups
# Apply a certain calculation 
# Combine the results back into a new data structure 
```

```python
# Find the annual mean for each numeric column. 
# Create a new column to aggregate on. - Release Year 
movies.Release_Date[19].year # Some arbitrary row
```

```python
def get_year(date):
    return date.year
```

```python
# Using the apply method is taking another function as an input
# and applys it to each value in a column or series. 
# Function being applied should take a single value as an input
# and produce a similar output which can be assigned to a new series. 

# Creating a Year column. 
movies["Year"] = movies.Release_Date.apply(get_year)
movies["Year"] = movies.Release_Date.apply(lambda x: x.year)

print(movies.Year.min())
print(movies.Year.max())
```

```python
movies.head()
```

```python
# groupby() method
by_year = movies.groupby('Year')
# After the groupby method is used it becomes a dictionary and the group
# labels (years) with the indices of all the rows that belong to that group.
print(by_year.groups.keys())
```

```python
# The indices are the row number. 
by_year.groups[1916]
```

```python
# Creating annual_means(df) which is using the mean()
# to calculate the mean of the grouped years indexed by year. 
annual_means = by_year.mean()
annual_means.info()
```

```python
annual_means.head()
```

```python
# How many movies released each year?
counts = by_year.aggregate(len)
counts.head()
```

```python
# Making count a series and storing the count of movies released per year.
counts = by_year.Movie.aggregate(len)
counts
```
