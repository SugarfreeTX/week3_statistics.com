```python
# Convert date
# locate unknown values in US Gross and Worldwide Gross column to 
# conver to 0.0. Also fill in missing values. 
def make_date(date_str):
    m, d, y = date_str.split('/')
    m = int(m)
    d = int(d)
    y = int(y)
    if y > 13:
        y += 1900
    else: 
        y += 2000
    return dt.datetime(y, m, d)

# Convert to int. 
def convert_to_int(text):
    try:
        return int(text) # convert text to int
    except:
        return None # if no text or value then return None
    
```

```python
# a. What is the median profit of movies with budget of over $50 mil.
# Need only budget over $50 mill 
# What do i need to do to get the median profit. 
# Create a profit column and then take only the one where budget is over
# $50mil and median(). 
```

```python
# Create a profit column but first need to clean up the data. 
# Find all unknown and replace with 0.0 if a column does not have data
# subsitite it with another columns data. 
# turn dates into date object. 
# convert budget, grosses into int. 
file.info()
```

```python
print(file["US Gross"].unique())
print(file["Worldwide Gross"].unique())
```

```python
movies = pd.read_csv("moviedata.csv", sep="\t", na_values = ['Unkno', 'Unknown'], 
                     parse_dates=[0], date_parser=make_date)
```

```python
movies.head()
```

```python
# Fill NaN with 0
movies.fillna(0, inplace=True) # Need the inplace 
```

```python
movies.head()
```

```python
# Create a profit column but first check for zero values in 
# worldwide gross. 
movies[movies["Worldwide Gross"]==0]
```

```python
movies["Worldwide Gross"][movies["Worldwide Gross"] == 0] = movies["US Gross"]
```

```python
# Create a profit column by subtracting Worldwide Gross from Budget.
movies["Profit"] = movies["Worldwide Gross"] - movies["Budget"]
```

```python
movies.head()
```

```python
# a. What is the median profit of movies with budgets of over $50M?
# Should I add profits where budget is over $50 mil and then median? 
# Create a new df/subset of the values we only one. 
bigger_budget = movies[movies["Budget"] > 50000000]
```

```python
bigger_budget["Profit"].median()
```

```python
# Find the number of movies released by each distributor.
# Use the groupby and aggregate method. 
# Create a subset and a subset of the previous subset. 
distributors = movies.groupby("Distributor").aggregate(len)
# Create subset, group the distributors all together and count. 
distributor_count = distributors["Movie"]
# subset created to take one column, any that will turn it into a Series. 
```

```python
distributor_count
# print the series 
```

```python
# output to csv with the to_csv()
distributor_count.to_csv("Distributor Counts.csv")
```

```python
# c. What are the mean and median movie profits by decade? 
# Which decade was the most profitable?
# Need to split off the year from the rest of the date and only take decade.
movies.head()
```

```python
def get_decade(date):
    year = date.year
    y = year % 10 # divide by and use remainder
    return year - y
# Example 1969 % 10 is 9. 1969-9 = 1960 shows decade
```

```python
movies["Decade"] = movies.Release_Date.apply(get_decade)
```

```python
movies.head()
```

```python
# # c. What are the mean and median movie profits by decade? 
# Which decade was the most profitable?
# make df with current movies but groupby the year 
decade_only = movies.groupby("Decade")
```

```python
decade_only
```

```python
# use mean and median to calculate profit by decade. 
decade_median = decade_only["Profit"].median()
decade_mean = decade_only["Profit"].mean()
```

```python
decades = pd.DataFrame({"Mean Profit": decade_mean, "Median Profit": decade_median})
decades
# Use the DataFrame method and create dictionaries with the values in place.
# no list brackets. 
```
